```python
from PIL import Image
import numpy as np
import os

def operation(img_path1, img_path2, operator, save_path):
    try:
        #open images and convert to 2D array of intensities
        with Image.open(img_path1) as img1:
            img_size = img1.size
            img_arr1 = np.array(img1, dtype=np.uint8)
        with Image.open(img_path2) as img2:
            img_arr2 = np.array(img2.resize(img_size), dtype=np.uint8)
        #initialise result array to 0
        result = np.zeros((img_size[1],img_size[0]), dtype=np.uint8)
        #perform operation
        for i in range(img_size[1]):
            for j in range(img_size[0]):
                result[i][j] = operator(img_arr1[i][j],img_arr2[i][j])
        #save result array to image
        result_img = Image.fromarray(result)
        result_img.save(save_path)
    except Exception as e:
        print(e)

def img_add(img_path1, img_path2):
    operator = lambda x,y: min(255,int(x)+int(y))
    operation(img_path1, img_path2, operator, 'add.tif')

def img_sub(img_path1, img_path2):
    operator = lambda x,y: abs(int(x)-int(y))
    operation(img_path1, img_path2, operator, 'sub.tif')

def img_mul(img_path1, img_path2):
    operator = lambda x,y: (int(x)*int(y))//255
    operation(img_path1, img_path2, operator, 'mul.tif')

def img_div(img_path1, img_path2):
    operator = lambda x,y: (int(x)/int(y) if y else 255)*255
    operation(img_path1, img_path2, operator, 'div.tif')

def img_and(img_path1, img_path2):
    operator = lambda x,y: (255 if x else 0)&(255 if y else 0)
    operation(img_path1, img_path2, operator, 'and.tif')

def img_or(img_path1, img_path2):
    operator = lambda x,y: (255 if x else 0)|(255 if y else 0)
    operation(img_path1, img_path2, operator, 'or.tif')

def img_xor(img_path1, img_path2):
    operator = lambda x,y: (255 if x else 0)^(255 if y else 0)
    operation(img_path1, img_path2, operator, 'xor.tif')

def img_not(img_path1):
    operator = lambda x,y: 0 if x else 255
    operation(img_path1, img_path1, operator, 'not.tif')

def img_and_not(img_path1, img_path2):
    operator = lambda x,y: (255 if x else 0)&(0 if y else 255)
    operation(img_path1, img_path2, operator, 'and_not.tif')

img_add(os.getcwd()+'/DIP3E_Original_Images_CH02/Fig0226(galaxy_pair_original).tif',os.getcwd()+'/DIP3E_Original_Images_CH02/Fig0236(a)(letter_T).tif')
img_sub(os.getcwd()+'/DIP3E_Original_Images_CH02/Fig0228(a)(angiography_mask_image).tif',os.getcwd()+'/DIP3E_Original_Images_CH02/Fig0228(b)(angiography_live_ image).tif')
img_mul(os.getcwd()+'/DIP3E_Original_Images_CH02/Fig0230(a)(dental_xray).tif',os.getcwd()+'/DIP3E_Original_Images_CH02/Fig0230(b)(dental_xray_mask).tif')
img_div(os.getcwd()+'/DIP3E_Original_Images_CH02/Fig0229(a)(tungsten_filament_shaded).tif',os.getcwd()+'/DIP3E_Original_Images_CH02/Fig0229(b)(tungsten_sensor_shading).tif')
img_and(os.getcwd()+'/DIP3E_Original_Images_CH02/Fig0230(b)(dental_xray_mask).tif',os.getcwd()+'/DIP3E_Original_Images_CH02/Fig0236(a)(letter_T).tif')
img_or(os.getcwd()+'/DIP3E_Original_Images_CH02/Fig0230(b)(dental_xray_mask).tif',os.getcwd()+'/DIP3E_Original_Images_CH02/Fig0236(a)(letter_T).tif')
img_xor(os.getcwd()+'/DIP3E_Original_Images_CH02/Fig0230(b)(dental_xray_mask).tif',os.getcwd()+'/DIP3E_Original_Images_CH02/Fig0236(a)(letter_T).tif')
img_and_not(os.getcwd()+'/DIP3E_Original_Images_CH02/Fig0230(b)(dental_xray_mask).tif',os.getcwd()+'/DIP3E_Original_Images_CH02/Fig0236(a)(letter_T).tif')
img_not(os.getcwd()+'/DIP3E_Original_Images_CH02/Fig0230(b)(dental_xray_mask).tif')
```
