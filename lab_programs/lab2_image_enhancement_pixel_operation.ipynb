```python
from PIL import Image
import numpy as np
import math,os

def negative(img_path, save_path):
    img = Image.open(img_path)
    img_shape=(img.width,img.height)
    img_array = np.array(img,dtype=np.uint8)
    for i in range(img_shape[1]):
        for j in range(img_shape[0]):
            img_array[i][j]=255-img_array[i][j]
    resultImage = Image.fromarray(img_array)
    resultImage.save(save_path)
    display(resultImage)

def logarithmic(img_path, save_path):
    c=2
    img = Image.open(img_path)
    img_shape=(img.width,img.height)
    img_array = np.array(img,dtype=np.uint8)
    img_array = img_array/255
    for i in range(img_shape[1]):
        for j in range(img_shape[0]):
            img_array[i][j] = c*math.log(img_array[i][j]+1)

    resultImage = Image.fromarray(img_array)
    resultImage.save(save_path)
    

def power(img_path, save_path):
    c=1
    gamma=5
    img = Image.open(img_path)
    img_shape=(img.width,img.height)
    img_array = np.array(img,dtype=np.uint8)
    img_array = img_array/255
    for i in range(img_shape[1]):
        for j in range(img_shape[0]):
            img_array[i][j] = c* (img_array[i][j]**gamma)

    resultImage = Image.fromarray(img_array)
    resultImage.save(save_path)
    

def contrast_stretching(img_path, save_path):
    img = Image.open(img_path)
    img_shape=(img.width,img.height)
    img_array = np.array(img,dtype=np.uint8)
    max = np.max(img_array)
    min = np.min(img_array)
    m = 255/(max-min)
    c = 255-m*max
    for i in range(img_shape[1]):
        for j in range(img_shape[0]):
            img_array[i][j] = m*img_array[i][j]+c

    resultImage = Image.fromarray(img_array)
    resultImage.save(save_path)
    display(resultImage)

negative(os.getcwd()+'/DIP3E_Original_Images_CH03/Fig0304(a)(breast_digital_Xray).tif','negative.tif')
logarithmic(os.getcwd()+'/DIP3E_Original_Images_CH03/Fig0304(a)(breast_digital_Xray).tif','log.tif')
power(os.getcwd()+'/DIP3E_Original_Images_CH03/Fig0309(a)(washed_out_aerial_image).tif','power.tif')
contrast_stretching(os.getcwd()+'/DIP3E_Original_Images_CH03/Fig0310(b)(washed_out_pollen_image).tif','contrast.tif')
```
