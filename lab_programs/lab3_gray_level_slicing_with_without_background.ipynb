```python
from PIL import Image
import numpy as np

def without_background(img_path, save_path):
    image = Image.open(img_path)
    img_array = np.array(image,dtype=np.uint8)
    img_shape=(image.width,image.height)
    
    display(image)
    
    for i in range(img_shape[1]):
        for j in range(img_shape[0]):
            if(img_array[i][j]>=0 and img_array[i][j]<=50):
                img_array[i][j]=255
            else:
                img_array[i][j]=0

    resultImage = Image.fromarray(img_array)
    resultImage.save(save_path)
    display(resultImage)

def with_background(img_path, save_path):
    image = Image.open(img_path)
    img_array = np.array(image,dtype=np.uint8)
    img_shape=(image.width,image.height)

    for i in range(img_shape[1]):
        for j in range(img_shape[0]):
            if(img_array[i][j]>=0 and img_array[i][j]<=50):
                img_array[i][j]=255

    resultImage = Image.fromarray(img_array)
    resultImage.save(save_path)
    display(resultImage)

with_background('Fig0222(b)(cameraman).tif','slicing_withbg.tif')
without_background('Fig0222(b)(cameraman).tif','slicing_withoutbg.tif')
```
